### Q. 1) Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.


```python
# HERE THERE WERE TWO NULL VALUES IN THE DATASETS IN CYLINDER COLUMN SO WE FILL THAT NULL VALUE WITH THE MEAN OF THE CYLINDER
# COLUMN AND HERE THE SYNTAX IS:---->

data['Cylinders'].fillna(data['Cylinders'].mean(),inplace=True)
```

### Q. 3) Instruction ( Filtering ) - Show all the records where Origin is Asia or Europe.

```python
# HERE WE USE THE VALUE_COUNT() SO THAT TO COUNT AND KNOW THE COUNTRY NAME BEFORE FILTERING THE DATASET
data['Origin'].value_counts()  
```

```python
# HERE WE MAKE A VARIABLE AND FILTER OUR DATA WITH THE HELP OF ISIN , AND THE REQUIREMENT WAS TO SHOW THE ORIGIN NAME AS ASIA 
# AND EUROPE WITHOUT USA.

filtering_data = data[data['Origin'].isin(['Asia','Europe'])] 
```

```python
filtering_data['Origin'].value_counts()
```

### Q. 4) Instruction ( Removing unwanted records ) - Remove all the records (rows) where Weight is above 4000.

```python
# HERE WE HAVE TO REMOVE THAT ROWS WHOSE WEIGHT IS GREATER THEN 4000 AND FOR REMOVING THAT ROWS THE SYNTAX IS WRITTEN BELOW.
# (~) THIS SIGN HELP US TO REMOVE ALL THE ROWS WHICH IS GRATER THEN 4000.
data[~(data['Weight']>4000)]
```

### Q. 5) Instruction ( Applying function on a column ) - Increase all the values of 'MPG_City' column by 3.


```python
# HERE WE INCREASE THE VALUE OF A PARTICULAR COLUMN WHICH NAME IS MPG_CITY BY 3 AND WE USE LAMBDA FUNCTION TO INCREASE THE VALUE
data['MPG_City']=data['MPG_City'].apply(lambda x:x+3)
```

```python
# THIS IS AN ANOTHER EXAMPLE OF LAMBDA FUNCTION AND IT SHOWS HOW WE CAN INCREASE AND DECREASE THE VALUE ON ANY COLUMN WITH THIS
# FUNCTION
data['MPG_Highway'] = data['MPG_Highway'].apply(lambda x:x+10)
```
